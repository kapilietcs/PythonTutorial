#### 2. Basic Line Plot
The **line** plot is one of the simplest and most common plots.

Example:
- Visualizing the growth of a function (e.g., y = x²):

```python
import numpy as np

# Data: x values and their corresponding y values (y = x²)
x = np.linspace(-10, 10, 100)
y = x ** 2

# Create a line plot
plt.figure(figsize=(8, 4))  # Set figure size (optional)
plt.plot(x, y, label="y = x²", color="blue", linestyle="--", linewidth=2)

# Add title and labels
plt.title("Line Plot Example: y = x²", fontsize=14)
plt.xlabel("x-axis", fontsize=12)
plt.ylabel("y-axis", fontsize=12)

# Add grid
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

**The Anatomy of a Matplotlib Plot**

The components of a typical Matplotlib figure:

1. **Figure**: The entire visualization.
2. **Axes**: The area where data is plotted (each figure can have multiple axes/subplots).
3. **Title, Labels, Legends**: Provide context to the visualization.
4. **Ticks**: The scale indicators on the x- and y-axes.
Here’s an example structure you’ll see often:

```python
plt.figure()                     # Create a blank canvas
plt.plot(x, y)                   # Add your plot (line chart, scatterplot, etc.)
plt.title("Title")               # Add a title
plt.xlabel("X-axis label")       # Label the x-axis
plt.ylabel("Y-axis label")       # Label the y-axis
ticks = np.arange(-10, 11, 1)    # Set x-ticks
plt.xticks(ticks)                # Set x-ticks   
plt.legend(["y = x²"])           # Add a legend
plt.show()                       # Display the plot
```

#### 3. Scatter Plot
Use **scatter plots** to show relationships between two variables.

- Example: Plot random data points:

```python
# Random Data
# np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y, c="red", marker="o", alpha=0.5, s=10, label="Random Points")

# Add title and labels
plt.title("Scatter Plot Example", fontsize=14)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()

# Show grid
plt.grid(True)

# Show the plot
plt.show()

```

#### 4. Bar Plot
Bar plots are great for comparing categories or groups.

- Example: Sales numbers for different products:

```python
# Categories and their values
categories = ['Product A', 'Product B', 'Product C']
sales = [120, 80, 150]

# Create a bar plot
plt.bar(categories, sales, color=['blue', 'green', 'orange'])

# Add title and labels
plt.title("Bar Plot Example: Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales (Units)")

# Show the plot
plt.show()
```

#### 5. Pie Chart
Pie Charts are great for showing the proportion of each category within a whole.

- Example: Sales numbers for different products:

```python
# Categories and their values
categories = ['Product A', 'Product B', 'Product C']
sales = [120, 80, 150]

# Create a pie chart
plt.pie(sales, labels=categories, autopct='%1.1f%%', colors=['blue', 'green', 'orange'], startangle=140)

# Add title
plt.title("Pie Chart Example: Product Sales Distribution")

# Equal aspect ratio ensures pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```

#### 6. Histogram
Histograms are used to visualize the distribution of data by grouping it into intervals (bins).

- Example: Visualizing the age distribution of a random population.

```python
# Generate random ages
ages = np.random.randint(10, 60, 100)

# Create a histogram
plt.hist(ages, bins=20, color="purple", edgecolor="black", alpha=0.7)

# Add title and labels
plt.title("Histogram Example: Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Count")

# Show the plot
plt.show()
```

#### 7. Customizing Your Matplotlib Plots
Once you’ve created a plot, it’s important to make it readable and visually appealing. Customizations allow you to communicate insights effectively.

**Adding a Title, Axis Labels, and Grid**

```python
plt.title("Title of the Plot", fontsize=16)
plt.xlabel("X-Axis Label", fontsize=12)
plt.ylabel("Y-Axis Label", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
```

**Legends**

Legends tell the reader what different elements in a plot represent. Use label when plotting, then call plt.legend().

```python
# Multiple lines with a legend
x = np.linspace(-2, 2, 50)
plt.plot(x, x ** 2, label="y = x²")
plt.plot(x, x ** 3, label="y = x³")
plt.legend(loc="upper left")
```

**Customizing Colors, Linestyles, and Markers**

```python
# Customizing the plot with colors, line styles, and markers
plt.title("Customizing the Plot", fontsize=16)
plt.plot(x, x ** 2, color="green", linestyle=":", marker="o")
```

- Colors: `'red', 'blue', 'green'`, or HEX codes (e.g., #ff5733).
- Linestyles: `'-', '--', '-.', ':'`.
- Markers: Shapes of points (e.g., `'o', 's', '*', '^'`).

**Subplots**
Sometimes you need to show multiple plots in one figure to compare trends.

```python
# Create subplots
plt.figure(figsize=(10, 5))

# First subplot
plt.subplot(2, 2, 1)  # (2 Row, 2 Columns, Plot 1)
plt.plot(x, x ** 2, color="blue")
plt.title("y = x²")

# Second subplot
plt.subplot(2, 2, 2)  # (2 Row, 2 Columns, Plot 2)
plt.plot(x, x ** 3, color="green")
plt.title("y = x³")

# Third subplot
plt.subplot(2, 2, 3)  # (2 Rows, 2 Columns, Plot 3)
plt.plot(x, np.sin(x), color="red")
plt.title("y = sin(x)")


plt.tight_layout()  # Adjust spacing
plt.show()
```

#### 7. Saving Your Plots
Once you’ve created the perfect plot, save it to a file for use in presentations or reports.

### Day 11 Exercise Solution

1. Explore the Titanic Dataset:

    - Use .describe() to summarize the numerical columns.
    - Calculate the percentage of passengers that survived and those that didn’t.

```python
import pandas as pd
# Load Titanic dataset
titanic_df = pd.read_csv("../data/titanic_dataset.csv")  # Replace with the path to your dataset
titanic_df.head()
```

```python
# 1. Use .describe() to summarize numerical columns
print("Summary of Numerical Columns:")
titanic_df.describe()
```

```python
# 2. Calculate the percentage of passengers that survived and those that didn’t
survival_counts = titanic_df['Survived'].value_counts(normalize=True) * 100

print("\nSurvival Percentage:")
print(f"Did Not Survive (0): {survival_counts[0]:.2f}%")
print(f"Survived (1): {survival_counts[1]:.2f}%")
```

2. Group By Analysis:

    - Group passengers by `Embarked` and calculate the total `Far`e collected from each embarkation point.
    - Group passengers by `Sex` and `Pclass`, and calculate the survival rate for each group.

```python
# 1. Group passengers by 'Embarked' and calculate total 'Fare' collected from each embarkation point
fare_by_embarked = titanic_df.groupby('Embarked')['Fare'].sum()
print("Total Fare Collected by Embarkation Point:\n", fare_by_embarked)
```

```python
# 2. Group passengers by 'sex' and 'pclass', and calculate the survival rate for each group
survival_rate = titanic_df.groupby(['Sex', 'Pclass'])['Survived'].mean() * 100
print("\nSurvival Rate by Sex and Pclass (%):\n", survival_rate)
```

3. Filter Grouped Data:

    - Find all groups (by `Pclass`) where the average passenger fare exceeds 50.
    - Filter and print all passengers who paid more than $100 and were older than 30.


```python
# 1. Find all groups (by 'Pclass') where the average Fare exceeds 50
avg_fare_by_pclass = titanic_df.groupby('Pclass')['Fare'].mean()
high_fare_classes = avg_fare_by_pclass[avg_fare_by_pclass > 50]
print("Pclass groups with average fare > $50:\n")
high_fare_classes

```

```python
# 2. Filter passengers who paid more than $100 and were older than 30
filtered_passengers = titanic_df[(titanic_df['Fare'] > 100) & (titanic_df['Age'] > 30)]
print("\nPassengers with fare > $100 and age > 30:\n")
filtered_passengers.head()
```
